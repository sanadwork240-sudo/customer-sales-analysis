```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('CSV_FILES/Cleaned_format.csv')
df
```

```python
top_locations = (
    df.groupby('location')['purchase amount (usd)']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_locations
```

```python
plt.figure(figsize=(9, 5))

sns.barplot(
    x=top_locations.values,
    y=top_locations.index
)

plt.title('Top 10 Locations by Total Revenue')
plt.xlabel('Total Revenue (USD)')
plt.ylabel('Location')

plt.show()
```

Revenue is relatively evenly distributed across the top-performing locations, with Montana contributing the highest sales. The small difference among the top 10 locations suggests that the business maintains a fairly balanced geographical sales distribution rather than depending heavily on a single location.
