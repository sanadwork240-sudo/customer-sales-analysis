AGE ANALYSIS


```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))

sns.histplot(df['Age'], bins=50, kde=True)
plt.title('Distribution of Customer Age')
plt.xlabel('Age')
plt.ylabel('Number of Customers')

plt.show()
```

```python
plt.figure(figsize=(8,2))

sns.boxplot(x=df['Age'])

plt.title('Box Plot of Age')

plt.show()
```

```python
bins = [17, 25, 35, 45, 55, 65, 70]

labels = ['18-25',
          '26-35',
          '36-45',
          '46-55',
          '56-65',
          '66-70']

df['Age Group'] = pd.cut(df['Age'],
                         bins=bins,
                         labels=labels)
```

```python
df['Age Group'].value_counts().sort_index()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='Age Group',
    data=df,
    order=['18-25','26-35','36-45','46-55','56-65','66-70']
)

plt.title('Customer Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Customers')

plt.show()
```

```python
df.columns = df.columns.str.lower().str.strip()
```

```python
df.columns
```

```python
avg_purchase = (
    df.groupby('age group')['purchase amount (usd)']
      .mean()
      .reset_index()
)

avg_purchase
```

```python
sns.barplot(
    x='age group',
    y='purchase amount (usd)',
    data=avg_purchase
)
```

```python
revenue = df.groupby('age group')['purchase amount (usd)'].sum().reset_index()
```

```python
sns.barplot(
    x='age group',
    y='purchase amount (usd)',
    data=revenue
)
```
