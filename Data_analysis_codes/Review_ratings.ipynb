```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('CSV_FILES/removed.csv')
df
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['review rating'],
    bins=10,
    kde=True
)

plt.title('Distribution of Review Ratings')
plt.xlabel('Review Rating')
plt.ylabel('Number of Customers')

plt.show()
```

```python
plt.figure(figsize=(8,2))

sns.boxplot(x=df['review rating'])

plt.title('Box Plot of Review Ratings')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='review rating',
    y='purchase amount (usd)',
    data=df,
    alpha=0.6
)

plt.title('Review Rating vs Purchase Amount')
plt.xlabel('Review Rating')
plt.ylabel('Purchase Amount (USD)')

plt.show()
```

There is no significant correlation between customer review ratings and purchase amount. Customers who spend more do not necessarily provide higher ratings, indicating that spending behavior is largely independent of customer satisfaction ratings in this dataset.
