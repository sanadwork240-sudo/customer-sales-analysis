```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(7, 5))

sns.countplot(
    x='category',
    data=df,
    order=df['category'].value_counts().index
)
plt.title('Number of Purchases by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Number of Purchases')

plt.show()
```

```python
category_revenue = (
    df.groupby('category')['purchase amount (usd)']
    .sum()
    .sort_values(ascending=False).reset_index()
)

category_revenue
```

```python
plt.figure(figsize=(7,5))
sns.barplot(
    x='category',
    y='purchase amount (usd)',
    data=category_revenue
)

plt.title('Total Revenue by Category')
plt.xlabel('Category')
plt.ylabel('Total Revenue (USD)')

plt.show()

```

```python
category_revenue_average = (
    df.groupby('category')['purchase amount (usd)']
    .mean()
    .sort_values(ascending=False).reset_index()
)

category_revenue_average
```

```python
plt.figure(figsize=(7,6))

sns.barplot(  x='category',
    y='purchase amount (usd)',
    data=category_revenue_average)


plt.xlabel('Category')
plt.title('Average Purchase Amount by Category')
plt.ylabel('Average Purchase Amount (USD)')

plt.show()
```

THE TOTAL REVENUE GENRATED IS MOST IN CASE OF THE CLOTHING THAN OTHERS BUT THE AVERAGE REVENUE GENRATED PER CUSTOMER IS SAME AS THAT OF THE OTHER CATEGORYS BUT IN CASE OF OUTWARE THE AVG IS IS A LITTLE DROP.

```python
category_rating = (
    df.groupby('category')['review rating']
    .mean()
    .sort_values(ascending=False)
)

category_rating
```

```python
plt.figure(figsize=(7, 5))

sns.barplot(
    x=category_rating.index,
    y=category_rating.values
)

plt.title('Average Review Rating by Category')
plt.xlabel('Category')
plt.ylabel('Average Review Rating')

plt.show()
```

```python
season_category = pd.crosstab(
    df['season'],
    df['category']
)

season_category
```

```python
plt.figure(figsize=(9, 5))

sns.countplot(
    x='season',
    hue='category',
    data=df
)

plt.title('Product Category Purchases by Season')
plt.xlabel('Season')
plt.ylabel('Number of Purchases')

plt.show()
```
