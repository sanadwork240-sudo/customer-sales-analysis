```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('CSV_FILES/Cleaned_format.csv')
df
```

```python
top_10_items = df['item purchased'].value_counts().head(10)

top_10_items
```

```python
plt.figure(figsize=(9, 5))

sns.barplot(
    x=top_10_items.values,
    y=top_10_items.index
)

plt.title('Top 10 Most Purchased Items')
plt.xlabel('Number of Purchases')
plt.ylabel('Item Purchased')

plt.show()
```

```python
top_10_item_revenue = (
    df.groupby('item purchased')['purchase amount (usd)']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_10_item_revenue
```

```python
plt.figure(figsize=(9, 5))

sns.barplot(
    x=top_10_item_revenue.values,
    y=top_10_item_revenue.index
)

plt.title('Top 10 Items by Total Revenue')
plt.xlabel('Total Revenue (USD)')
plt.ylabel('Item Purchased')

plt.show()
```

```python
season_revenue = (
    df.groupby('season')['purchase amount (usd)']
    .sum()
    .sort_values(ascending=False)
)

season_revenue
```

```python
plt.figure(figsize=(7, 5))

sns.barplot(
    x=season_revenue.index,
    y=season_revenue.values
)

plt.title('Total Revenue by Season')
plt.xlabel('Season')
plt.ylabel('Total Revenue (USD)')

plt.show()
```

Product-level analysis shows that Blouse leads both purchase volume and total revenue. While several highly purchased items also appear among the top revenue-generating products, the rankings vary slightly, indicating differences in purchase value across products. Seasonal revenue remains relatively stable, with Fall generating the highest revenue and Summer recording the lowest.
